```python
pais = ['Colombia','Ecuador','Honduras','México','Perú','Promedio Regional']
td = [3.94,3.17,3.75,3.67,3.85,3.68]
""" Create lists of the state names and populations for the top 6 populus states """
tdpais = [(pais[i],td[i]) for i in range(len(pais))]
tdpais.sort(key = lambda x:x[1], reverse=True)
paisplot = [x[0] for x in tdpais]  
tdPlot = [x[1] for x in tdpais] 


fig,ax = plt.subplots()
fig.set_size_inches(14,6)
fig.suptitle("Score by country",fontsize='18',fontname='Times New Roman')
ax.bar(paisplot,tdPlot)
plt.show()
```
